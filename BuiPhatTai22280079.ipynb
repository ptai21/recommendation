1. The probability of passing three candidates A, B, and C in one interview are 60%,
10%, and 20% respectively. Find the probability of success of at least two (2 points)

```python
# xac suat lan luot cua cac ung cu vien dau buoi phong van 
# moi ung vien co ti le dau phong van doc lap voi nhau
p_A = 0.6
p_B = 0.1
p_C = 0.2
# xac suat khong co ung cu vien nao dau buoi phong van 
p_0 =(1-p_A)*(1-p_B)*(1-p_C)
# xac suat co 1 ung cu vien dau buoi phong van 
p_1 = (p_A)*(1-p_B)*(1-p_C) + (1-p_A)*p_B*(1-p_C) +(1-p_A)*(1-p_B)*(p_C)
# xac suat co it nhat 2 ung cu vien dau buoi phong van 
p_result = 1 - p_0 - p_1
print(f"xac suat co it nhat 2 ung cu vien dau buoi phong van: {p_result}")
```

2. Nine cards are drawn from a deck of 52 cards. What is the probability that 5 cards 
are diamonds and 2 are hearts and 2 are kings? (2 points)

```python
# so cach rut duoc 5 la bai ro 2 la bai co va 2 la king 
# nCk la to hop chap k cua n 
# tong so cach rut la 
total_way = comb(52,9)
# th1: 5 la bai ro va 2 la bai co khong chua king 
case_1 = comb(12,5)*comb(12,2)
#th2: 4 la bai ro va co 1 la k ro va 2 la co cong la 1 la k (chuon hoac bich) va 1 la bai chuong hoac bich (khac K)
case_2 = comb(12,4)*comb(12,2)*comb(2,1)*comb(24,1)
#th3 nguoc lai ti voi th2 la co 5 quan ro khac K va co mot quan k co
case_3 = comb(12,5)*comb(12,1)*comb(2,1)*comb(24,1)
#th4 hai quan k deu la ro va co 
case_4 = comb(12,4)*comb(12,1)*comb(24,2)
#
result = (case_1+case_2+case_3+case_4)/total_way
print(f"Xac suat rut bai duoc la: {result}")
```

3. There are transactions from banks A, B, and C at your company. About 30% of the 
population comes from A, 20% from B, and 50% from C. Knowing that there is  
chance of fraud in the transaction history. The percentages for banks A,  , and C
are 3%, 5%, and 4%, respective y.
Which bank will most likely belong in the event of fraud detection? (2 points)

```python
# phan bo dan so tham gia tai cac ngan hang 
p_A= 0.3
p_B = 0.2
p_C = 0.5
#ti le phan tram bi loi cua tung ngan hang
p_fraud_A = 0.03 # p(froud|A)
p_fraud_B = 0.05 # p(froud|B)
p_fraud_C = 0.04 # p(froud|C)
# xac suat gian luan ap dung xac suat toan phan 
p_fraud = p_fraud_A*p_A + p_fraud_B*p_B + p_fraud_C*p_C
# ti le phat hein gian lan cua tung ngan hang
# ngan hang A p(A|fraud)
p_A_fraud = (p_fraud_A*p_A)/p_fraud
p_B_fraud = (p_fraud_B*p_B)/p_fraud
p_C_fraud = (p_fraud_C*p_C)/p_fraud
print(f"phan tram phat hien cua ngan hang A: {p_A_fraud*100:.2f}")
print(f"phan tram phat hien cua ngan hang B: {p_B_fraud*100:.2f}")
print(f"phan tram phat hien cua ngan hang C: {p_C_fraud*100:.2f}")
if p_A_fraud - p_B_fraud > 0 :
    if p_A_fraud - p_C_fraud > 0:
        print(f"-> Ngan hang A co ti le phat hien cao nhat: {p_A_fraud*100:.2f}")
    else:
        print(f"-> Ngan hang C co ti le phat hien cao nhat: {p_C_fraud*100:.2f}")
else:
    if p_B_fraud - p_C_fraud > 0:
        print(f"-> Ngan hang B co ti le phat hien cao nhat: {p_B_fraud*100:.2f}")
    else:
        print(f"-> Ngan hang C co ti le phat hien cao nhat: {p_C_fraud*100:.2f}")
  
```

4. Assume that the amount of transaction in Service X in your company follow the 
Normal distribution with mean ¬µ. A random sample of 250 000 transactions yielde 
an average amount of ùë• ÃÖ= 525 000 VND, and the standard deviation of the 250
000 transactions is found to be s = 28 340. A 90% confidence interval for ¬µ is (Hi t:
The t critical value for 90% CI for Degrees of freedom = 24 is 1.711) (2 points)nts)

```python
# tim khoan tin cay 90% cho u  Truong hop khong biet phuong sai
n = 250000
x_mean = 525000
s = 28340
Z_alpha_2 = 1.645 # truong hop khong biet phuong sai va co mau lon 
u1 = x_mean - Z_alpha_2*s/sqrt(n)
u2 = x_mean + Z_alpha_2*s/sqrt(n)
print(f"khoang tin cay 90% cua u la: [{u1:.2f}:{u2:.2f}]")
```

# ***Feature Engineering***

M·ª•c ƒë√≠ch x√¢y d·ª±ng m·ªôt m√¥ h√¨nh ph√¢n lo·∫°i ƒë·ªÉ c√≥ th·ªÉ Recommendation System cho n√™n gi√° tr·ªã ban ƒë·∫ßu ch∆∞a ph√π h·ª£p l·∫Øm cho n√™n c√≥ th·ªÉ l·ªçc l·∫°i v√† gh√©p c√°c d·ªØ li·ªáu c·∫ßn thi·∫øt v·ªõi m·ª•c ƒë√≠ch

```python
#l·∫•y ra t·ªïng s·ªë transactionId c·ªßa m·ªói User
total_transactions = data_1.groupby('UserId')['TransactionId'].nunique().reset_index()
total_transactions.rename(columns={'TransactionId': 'Total_Transactions'}, inplace=True)
#l·∫•y ra t·ªïng s·∫£n ph·∫©m mua c√°c lo·∫°i h√†ng 
data_1['TransactionTime'] = pd.to_datetime(data_1['TransactionTime'], errors='coerce')
unique_products_purchased = data_1.groupby('UserId')['ItemCode'].nunique().reset_index()
unique_products_purchased.rename(columns={'ItemCode': 'Unique_Products_Purchased'}, inplace=True)
# l·∫•y ra s·ªë ng√†y mua c≈©n nh∆∞ th·ªùi gian mua 
data_1['Day_Of_Week'] = data_1['TransactionTime'].dt.dayofweek
data_1['Hour'] = data_1['TransactionTime'].dt.hour
# Th·ªùi gian gi·ªØa hai l·∫ßn mua li·ªÅn ti·∫øp
days_between_purchases = data_1.groupby('UserId')['TransactionTime'].apply(lambda x: (x.sort_values().diff().dropna()).apply(lambda y: y.days))
avg_days_between_purchases = days_between_purchases.groupby('UserId').mean().reset_index()
avg_days_between_purchases.rename(columns={'TransactionTime': 'Average_Days_Between_Purchases'}, inplace=True)
# ng√†y mua trong tu·∫ßn
shopping_day = data_1.groupby(['UserId', 'Day_Of_Week']).size().reset_index(name='Count')
shopping_day = shopping_day.loc[shopping_day.groupby('UserId')['Count'].idxmax()][['UserId', 'Day_Of_Week']]
# gi·ªù mua trong ng√†y
shopping_hour = data_1.groupby(['UserId', 'Hour']).size().reset_index(name='Count')
shopping_hour = shopping_hour.loc[shopping_hour.groupby('UserId')['Count'].idxmax()][['UserId', 'Hour']]
# gh√©p c√°c k√©t qu·∫£ l·ªçc gom v√†o data frame ban ƒë·∫ßu
data_1 = pd.merge(data_1, total_transactions, on='UserId', how='left')
data_1 = pd.merge(data_1, unique_products_purchased,  on='UserId', how='left')
data_1 = pd.merge(data_1, avg_days_between_purchases ,on='UserId', how='left')
data_1 = pd.merge(data_1, shopping_day ,on='UserId', how='left')
data_1 = pd.merge(data_1, shopping_hour ,on='UserId', how='left')
```

```python
pca = PCA(random_state= 42)
X_pca = pca.fit_transform(X_scale_temp) 
# v·∫Ω bi·ªÉu ƒë·ªì th·ªÉ hi·ªán t·ªâ l·ªá ph∆∞∆°ng sai ƒë∆∞·ª£c gi·∫£i th√≠ch
plt.figure(figsize=(8, 4))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, len(np.cumsum(pca.explained_variance_ratio_)) + 1), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('T·ª∑ L·ªá Ph∆∞∆°ng Sai gi·∫£i th√≠ch ƒë∆∞·ª£c')
plt.xlabel('S·ªë Components')
plt.title('Ph∆∞∆°ng sai ƒë∆∞·ª£c gi·∫£i th√≠ch b·ªüi c√°c th√†nh ph·∫ßn ch√≠nh kh√°c nhau')
plt.tight_layout()
plt.show()
# ch·ªçn s·ªë components ƒë∆∞·ª£c gi·ªØ l·∫°i d·ª±a tr√™n ph∆∞∆°ng sai 
var_thres = 0.90
cumulative_var = np.cumsum(pca.explained_variance_ratio_)
n_components = np.where(cumulative_var >= var_thres)[0][0] + 1
# gi·∫£m chi·ªÅu 
print('s·ªë components ƒë∆∞·ª£c gi·ªØ l·∫°i',n_components)
```
